```python
"""tarea2"""
cateto_opuesto = int(input("Ingrerse el cateto opuesto(solo numeros enteros)"))
cateto_adyacente = int(input("Ingrerse el cateto adyacente(solo numeros enteros)"))

def calcular_hipotenusa(cateto_opuesto, cateto_adyacente):
    hipotenusa = (cateto_opuesto**2 + cateto_adyacente**2)**0.5
    return hipotenusa

hipotenusa = calcular_hipotenusa(cateto_opuesto, cateto_adyacente)

print(f"la hipotenusa es: {hipotenusa}")
```

```python
"""tarea"""
import time


productos = {"vestido de seda": 100, "Camiseta edicion especial de bob esponja": 3000, "Zapatos de cuero": 500, "Sombrero de vaquero": 450}
inventario = {"vestido de seda": "a01", 
              "Camiseta edicion especial de bob esponja": "b02",
                "Zapatos de cuero": "c03",
                  "Sombrero de vaquero": "d04"}
carrito = []
recibo = []
bolsa= []


print("Bienvenido a la tienda de Pedro")
time.sleep(1)
print(

    "vestido de seda\n   codigo de producto: a01\n   Precio: $100\n"
    "Camiseta edicion especial de bob esponja\n   Codigo del producto: b02\n   Precio: $3000\n"
    "Zapatos de cuero\n   Codigo del producto: c03\n   Precio: $500\n"
    "Sombrero de vaquero\n   Codigo del producto: d04\n   Precio: $450\n"
)

a01 = "vestido de seda"
b02 = "Camiseta edicion especial de bob esponja"
c03 = "Zapatos de cuero"
d04 = "Sombrero de vaquero"
cantidad_a01 = carrito.count(a01)
cantidad_b02 = carrito.count(b02)
cantidad_c03 = carrito.count(c03)
cantidad_d04 = carrito.count(d04)


while True:
    carrito_principal = input("Ingrese un producto (o escriba X para terminar): ")
    
    if carrito_principal == "X":
        break
    elif carrito_principal in productos:
        carrito.append(carrito_principal)
        print(f"{carrito_principal} agregado al carrito")

for producto in carrito:
    bolsa.append(inventario[producto])   
    recibo.append(productos[producto])

def calcular_total(lista_precios):
    return sum(lista_precios)

total = calcular_total(recibo)

suma_recibo = " + ".join(str(precio) for precio in recibo)
print(f"Productos en bolsa: {bolsa}\nPrecios en recibo: {suma_recibo} = ${total}")
```
